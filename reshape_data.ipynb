```python
import pandas as pd
import numpy as np
```

```python
# Read factor data in and get data for specified year
data = pd.read_csv('insample.csv')
# data = data.query("year==" + str(2009))

data.head()
```

```python
table = pd.pivot_table(data, values=['sale_invcap', 'roa', 'GProf'], index='year', columns='permno')
table
```

```python
# Is mean value because pivot_table took mean over the year as default aggregation function?
applied = table.apply(lambda x: x / x.shift(1)) # This year's mean value over last year's value
```

```python
test = applied.loc[2007, (slice(None), 10002)]
test
```

```python

weights = {
    'GProf': 1.9,
    'roa': 0.3,
    'sale_invcap': 1.4,
}
results = {}
score = 0
x = 0
vals = test.values
why_not = []
fund_mom = np.array([])
def fund_mom_score(year: int, permno: int, weights: dict):
    """Takes a year, permno, and dict with weights for each fundamental factor, then computes a 
    fundamental momentum score representing a combination of the year to year trend in those fundamental factors
    for that specific ticker for the specific year.
    """
    year_ticker_slice = applied.loc[year, (slice(None), permno)]
    vals = year_ticker_slice.values
    score = 0
    for i, factor in enumerate(year_ticker_slice.keys()):
        results[factor[0]] = vals[i]
    for key in results.keys():
        score = score + (weights[key] * results[key])
    return score
for year, permno in zip(data['year'].values, data['permno'].values):
    f_score = fund_mom_score(year, permno, weights)
    np.append(fund_mom, f_score)
    why_not.append(f_score)
#     print(fund_mom_score(year,permno,weights))
    
# np.append(fund_mom, fund_mom_score(2007, 10002, weights))

# Idea, combine these values into one "general fundamental momentum" value
# Each company should have a fund_mom value for that year
# i.e. each permno should get a fund_mom value added
# We're adding a fund_mom column
```

```python
fund_mom
```

```python
len(why_not)
```

```python
data['fund_mom'] = pd.Series(why_not)
```

```python
data.shape
```

```python
data.to_csv('./in_sample_w_fund_mom.csv')
```
