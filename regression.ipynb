```python
yearly_data_t = []
yearly_data_coef = []
total_df = pd.DataFrame([], columns = all_factors)
for year in range(1981,2008):
    print(year)
    raw_df = data_dict[year]
    tempdf=pd.DataFrame([])
    #zscore the factor
    for factor_name in all_factors:
        tempdf[factor_name]=raw_df.groupby(['INDUSTRY'])[factor_name].transform(lambda x: (x - x.mean()) / x.std())
    
    #set over and below 3 std to 3 std.
    #There are many ways to deal with outliers, this is just an example
    np_matrix = tempdf[all_factors].values
    nrow = np_matrix.shape[0]
    ncol = np_matrix.shape[1]
    for irow in range(nrow):
        for icol in range(ncol):
            if np.isnan(np_matrix[irow,icol]):
                np_matrix[irow,icol] = 0
            elif np_matrix[irow,icol] > 3:
                np_matrix[irow,icol] = 3
            elif np_matrix[irow,icol] < -3:
                np_matrix[irow,icol] = -3
            else:
                pass
    factor_df = pd.DataFrame(np_matrix, columns = all_factors, index = raw_df.index)
    total_df = total_df.append(factor_df)
    
    X = sm.add_constant(factor_df.to_numpy())
    model = sm.OLS(raw_df[['ret1yrafter']],X)
    model = model.fit()
    
    yearly_data_t.append(model.tvalues.values)
    yearly_data_coef.append(model.params.values)
```

```python
columns = ['constant'] + all_factors
result_t = pd.DataFrame(yearly_data_t, columns = columns, index = range(1981,2008))
result_coef = pd.DataFrame(yearly_data_coef, columns = columns, index = range(1981,2008))
FM_stat = result_coef.mean()/result_coef.std()*np.sqrt(len(result_coef))

FM_stat.to_csv('FM_stat.csv')
result_coef.to_csv('coef.csv')
result_t.to_csv('t-statistics table.csv')

FM_stat.head()
```
